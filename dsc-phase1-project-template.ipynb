```python
def remove_non_digit_chars(string):
    return float(''.join(c for c in string if c.isdigit()))
```

```python
remove_non_digit_chars('$425,000,000')
```

```python
dfs['tn_movie_budgets']
```

```python
dfs['tn_movie_budgets'].production_budget = dfs['tn_movie_budgets'].production_budget.map(lambda x: remove_non_digit_chars(x))
```

```python
dfs['tn_movie_budgets'].worldwide_gross = dfs['tn_movie_budgets'].worldwide_gross.map(lambda x: remove_non_digit_chars(x))
```

```python
dfs['tn_movie_budgets']['wwgross_over_budget'] = dfs['tn_movie_budgets'].worldwide_gross / dfs['tn_movie_budgets'].production_budget
```

```python
dfs['tn_movie_budgets'].sort_values(by='wwgross_over_budget', ascending=False)
```
